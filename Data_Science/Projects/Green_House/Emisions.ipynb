```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 

data = pd.read_csv("global_greenhouse_gas_emission_from_agriculture.csv") 
data
```

```python

data = data.drop(columns=["Flag", "Note", "Item", "Unit", "Domain"])
data
```

```python
sanitized_names = data.columns.str.replace(" ", "_")

data.columns = sanitized_names
```

```python
data = data.drop(columns=["Flag_Description", "Domain_Code", "Source_Code", "Area", "Source"])
data
```

```python
def map_eletemt_to_number (ele:str) -> int:
    match ele:
        case "Emissions (CH4)": return 1
        case "Emissions (CO2)": return 2
        case "Emissions (N2O)": return 3
    
    return -1

data["Element"] = data["Element"].apply(map_eletemt_to_number)

```

```python
data = data.drop(columns=["Element"])
```
